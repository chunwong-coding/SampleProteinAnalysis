```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
deve_df = pd.read_csv("Resources/data1_by_panels/data1_DEVELOPMENT.csv")
deve_df = deve_df.set_index("SampleID")
deve_df
```

```python
treated = deve_df[deve_df["Treatment"]=="Treated"]
untreated = deve_df[deve_df["Treatment"]=="Untreated"]
```

```python

treated_b = treated.loc[treated["Time"]=="Baseline"]
treated_b_cleaned = treated_b.iloc[:, 0:93]
treated_s = treated.loc[treated["Time"]=="Week.6"]
treated_s_cleaned = treated_b.iloc[:, 0:93]
treated_t = treated.loc[treated["Time"]=="Week.12"]
treated_t_cleaned = treated_b.iloc[:, 0:93]
treated_cols = treated.iloc[:, 0:92].columns
```

```python
untreated_b = untreated.loc[untreated["Time"]=="Baseline"]
untreated_b_cleaned = untreated_b.iloc[:, 0:93]
untreated_s = untreated.loc[untreated["Time"]=="Week.6"]
untreated_s_cleaned = untreated_b.iloc[:, 0:93]
untreated_t = untreated.loc[untreated["Time"]=="Week.12"]
untreated_t_cleaned = untreated_b.iloc[:, 0:93]
untreated_cols = untreated.iloc[:, 0:92].columns
```

```python
treated
```

```python
x = [0, 1, 2]
x_labels = ["Baseline", "Week 6", "Week 12"]
treated_points = []
untreated_points = []
for y in range(len(treated_b[treated_cols[0]])):
    point1 = treated_b[treated_cols[0]][y]
    point2 = treated_s[treated_cols[0]][y]
    point3 = treated_t[treated_cols[0]][y]
    points = [point1, point2, point3]
    plt.plot(x, points, "b", label="Treated")
for y in range(len(untreated_b[treated_cols[0]])):
    point1 = untreated_b[untreated_cols[0]][y]
    point2 = untreated_s[untreated_cols[0]][y]
    point3 = untreated_t[untreated_cols[0]][y]
    points = [point1, point2, point3]
    plt.plot(x, points, "r", label="Untreated")
plt.xticks(x, x_labels)
handles, labels = plt.gca().get_legend_handles_labels()
newLabels, newHandles = [], []
for handle, label in zip(handles, labels):
      if label not in newLabels:
        newLabels.append(label)
        newHandles.append(handle)
plt.title(treated_cols[0])
plt.legend(newHandles, newLabels, loc="best")
plt.show()
```
