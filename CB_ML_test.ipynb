```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Merge data sets

```python
df = pd.read_csv("Resources/NPX_data1_merged.csv")
df
```

```python
baseline_df = df.loc[df["Time"]=="Baseline"]
baseline_df.head()
```

```python
baseline_df_clean = baseline_df.drop(["SampleID", "Subject", "Treatment", "Site", "Time"], axis=1)
baseline_df_clean.head()
```

```python
column_names = baseline_df_clean.columns.tolist()
column_names
```

## K Means

```python

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)

# Fit the model to the data
kmeans.fit(baseline_df_clean)


# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters = kmeans.predict(baseline_df_clean)
print(predicted_clusters)
print(baseline_df_clean.shape)
print(type(predicted_clusters))
```

```python
# Plot the clusters
X = list(baseline_df_clean.index.values)
plt.scatter(X, predicted_clusters, c=predicted_clusters, s=10, cmap='viridis')
plt.show()
```

```python
# calculate distortion for a range of number of cluster
distortions = []
for i in range(1, 10):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(baseline_df_clean)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 10), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
