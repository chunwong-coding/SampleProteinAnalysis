```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Merge data sets

```python
file = "NPX_data2.xlsx"
df = pd.read_excel(file, skiprows=1)
```

```python
df_cleaned = df.set_index("NPX Data")
```

```python
#Drop null value rows
df_cleaned = df_cleaned.dropna(how='all')
df_cleaned
```

```python
df_cleaned = df_cleaned.iloc[:,0:1104]
df_cleaned
```

```python
col_len = len(df_cleaned.columns)
for x in range(col_len):
    df_cleaned.iloc[1][x] = f'{df_cleaned.iloc[1][x]}_{df_cleaned.iloc[3][x]}'
df_cleaned
```

```python
#Reset Index Values
df_cleaned = df_cleaned.reset_index()
#Rename column to SampleID for later merge
df_cleaned = df_cleaned.rename(columns={"NPX Data":"SampleID"})
df_cleaned
```

```python
#Drop unnecessary data rows
df_cleaned = df_cleaned.drop([2,3,164,165])
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.drop([162,163,164,165,166,167,168,169,170,171,172,173,174,175,176,177])
df_cleaned
```

```python
#Make new dataframe using Uniprot ID as headers
headers=df_cleaned.iloc[1]
new_df = pd.DataFrame(df_cleaned.values, columns=headers)
new_df.columns
```

```python
new_df= new_df.reset_index(drop=True)
```

```python
new_df
```

```python
#Rename and set the index to the Samples column
new_df = new_df.rename(columns={"Assay":"SampleID"})
new_df = new_df.drop([0,1])
new_df = new_df.set_index("SampleID")
```

```python
new_df
```

```python
#Read sample info data excel to dataframe
info_df = pd.read_excel("NPX_data2_info.xlsx")

info_df.head()
```

```python
#Merge dataframes together on SampleID
merged_df = new_df.merge(info_df, on="SampleID", how="inner")
merged_df
```

```python
merged_df = merged_df.drop(columns=["Project"])
```

```python
merged_df = merged_df.rename(columns={"Treatment":"Response"})
merged_df.loc[merged_df["Response"]=="Untreated","Response"] = "Nonresponder"
merged_df.loc[merged_df["Response"]=="Treated","Response"] = "Responder"
merged_df
```

```python
#Save data to CSV
merged_df.to_csv("Resources/NPX_data2_merged.csv", index=False)
```

```python
#Make dataframes for the different timeframes of data
baseline_df = merged_df.loc[merged_df['Time']=="Baseline"]
week6_df = merged_df.loc[merged_df['Time']=="Week.6"]
week12_df = merged_df.loc[merged_df['Time']=="Week.12"]
baseline_df
```

## Make scatter plots

```python
#Assign X values as the Sample IDs
bx1 = baseline_df.iloc[:,0]

#Set the Y values for the first protein ID from the different timeframe DFs
by1 = baseline_df.iloc[:,1].astype(float)
by2 = week6_df.iloc[:,1].astype(float)
by3 = week12_df.iloc[:,1].astype(float)
print(bx1.shape, by1.shape, by2.shape, by3.shape)
```

```python
print(bx1.values)
print(by1.values)
print(by2.values)
print(by3.values)
```

```python
#Make a scatter plot for the first protein
plt.scatter(bx1, by1, c='r', s=20)
plt.scatter(bx1, by2, c='g', s=20)
plt.scatter(bx1, by3, c='b', s=20)
plt.legend(('Baseline','Week 6','Week 12'))
plt.title("NPX for Protein O00553 over time")
plt.xlabel("Sample ID")
plt.xticks(rotation=90)
plt.ylabel("NPX")
plt.savefig("Protein_O00553_Sample.png")
```
