```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
organ_damage_df = pd.read_csv("Resources/data1_by_panels/data1_ORGAN DAMAGE.csv")
organ_damage_df = organ_damage_df.set_index("SampleID")
od_baseline = organ_damage_df.loc[organ_damage_df["Time"]=="Baseline"]
```

```python
x = od_baseline.columns
arrayO = []
for y in range(92):
    arrayO.append(x[y])
a = np.array(arrayO)
a.shape = (1,92)
```

```python
sites = od_baseline["Site"].unique()
sites_data = {
    "Site_A": [],
    "Site_B": [],
    "Site_C": [],
    "Site_D": [],
    "Site_E": []
}
```

```python
for xxx in range(len(sites)):
    for yyy in range(len(od_baseline)):
        #print(od_baseline.iloc[yyy]["Site"])
        if od_baseline.iloc[yyy]["Site"] == sites[xxx]:
            arrayOD = od_baseline.iloc[yyy, 0:92]
            sites_data[sites[xxx]].append(arrayOD)
```

```python
for x in range(len(sites)):
    if sites[x] == "Site_A":
        for ggg in range(len(sites_data[sites[x]])):
            plt.scatter(arrayO, sites_data[sites[x]][ggg], c="b", label="Site A", s=0.5)
            plt.xticks(fontsize=2, rotation=45)
    if sites[x] == "Site_B":
        for ggg in range(len(sites_data[sites[x]])):
            plt.scatter(arrayO, sites_data[sites[x]][ggg], c="r", label="Site B", s=0.5)
            plt.xticks(fontsize=2, rotation=45)
    if sites[x] == "Site_C":
        for ggg in range(len(sites_data[sites[x]])):
            plt.scatter(arrayO, sites_data[sites[x]][ggg], c="y", label="Site C", s=0.5)
            plt.xticks(fontsize=2, rotation=45)
    if sites[x] == "Site_D":
        for ggg in range(len(sites_data[sites[x]])):
            plt.scatter(arrayO, sites_data[sites[x]][ggg], c="g", label="Site D", s=0.5)
            plt.xticks(fontsize=2, rotation=45)
    if sites[x] == "Site_E":
        for ggg in range(len(sites_data[sites[x]])):
            plt.scatter(arrayO, sites_data[sites[x]][ggg], c="m", label="Site E", s=0.5)
            plt.xticks(fontsize=2, rotation=45)
plt.title("Organ Damage")
plt.show()
```
